1. Find Unique words in a sentence

```python
sentence_text = "python is easy and python is powerful"
words_list = sentence_text.split()  # Split the sentence into individual words
unique_words = set(words_list)      # Remove duplicates by converting to a set
print(f"Total unique words: {len(unique_words)}")
print(f"Unique words are: {unique_words}")
```

2. Employee with the Highest Salary

```python
employee_salaries = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
# Find the highest salary
max_salary = max(employee_salaries.values())        
# Find the highest salary employee
top_earner = max(employee_salaries)
print(f"Top Earning Employee: {top_earner} - {max_salary}")
```

3. Maximum and Minimum Numbers in a List

```python
numbers_list = [45, 22, 89, 10, 66]
#find max number
largest = max(numbers_list)
#find min number
smallest = min(numbers_list)
print(f"Numbers: {numbers_list}")
print(f"Largest number: {largest}")
print(f"Smallest number: {smallest}")
```

4. Count Products Above a Price Threshold

```python
product_prices = [450, 1200, 899, 1500, 300]
# we take threshold as 1000
threshold = 1000
# count how many oroducts greater than 1000
above_threshold = len([price for price in product_prices if price > threshold])
print(f"Products costing more than {threshold}: {above_threshold}")
```

5. Calculate Attendance Percentage

```python
attendance_record = ["P", "P", "A", "P", "P"]
# count number of days present
day_present = attendance_record.count("P")
#count total days
presentdays = len(attendance_record)
# calculate attendance percentage
attendance_percentage = (day_present /presentdays)*100
print(f"Attendance Percentage: {attendance_percentage:.2f}%")
```

6. Remove Duplicate Phone Numbers

```python
phone_numbers = [9876543210, 9123456789, 9876543210]
#remove duplicates by converting to a set
unique_numbers = set(phone_numbers)
print(f"Unique Phone Numbers: {unique_numbers}")
```

7. Count Character Frequency

```python
sample_string = "hello world"
dictionary = {}
for char in sample_string:
    dictionary[char] = dictionary.get(char, 0)+1
print(f"Character counts:{dictionary}")
```
