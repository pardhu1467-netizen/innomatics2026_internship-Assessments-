```python
# 1. Smart Light Controller
# create class to store status
class SmartLight:
    def __init__(self, light_name, current_state):
        self.light_name = light_name
        self.current_state = current_state.upper()

    def show_status(self):
        print(f"{self.light_name} is {self.current_state}")


# Sample Input
light1 = SmartLight("Bedroom Light", "ON")
light1.show_status()
```

```python
# 2: Employee ID Card System

# create class for storing employee details
class EmployeeCard:
    def __init__(self, name, emp_id, dept):
        self.name = name
        self.emp_id = emp_id
        self.dept = dept

    def display_details(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.dept}")


# Sample Input
emp = EmployeeCard("Rahul", "EMP102", "IT")
emp.display_details()
```

```python
# 3: Mobile Contact Record

#create class for storing mobile contact details
class Contact:
    def __init__(self, person_name, phone):
        self.person_name = person_name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.person_name}")
        print(f"Phone: {self.phone}")


# Sample Input
contact1 = Contact("Anita", 9876543210)
contact1.display_contact()
```

```python
# 4: Product Price Tag Generator

#create class for storing product details
class ProductTag:
    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price = price

    def print_tag(self):
        print(f"Product: {self.product_name}")
        print(f"Price: ₹{self.price}")


# Sample Input
item = ProductTag("Headphones", 2499)
item.print_tag()
```

```python
# 5: Movie Rating Display System

#create class for storing movie ratings
class MovieInfo:
    def __init__(self, title, rating):
        self.title = title
        self.rating = rating

    def show_movie(self):
        print(f"Movie: {self.title}")
        print(f"Rating: {self.rating} / 5")


# Sample Input
movie = MovieInfo("Inception", 4.8)
movie.show_movie()
```

```python
# 6: Delivery Address Manager

#create class for storing delivery details of the customer
class DeliveryInfo:
    def __init__(self, customer, location):
        self.customer = customer
        self.location = location

    def display_delivery(self):
        print("Delivery Details")
        print(f"Customer: {self.customer}")
        print(f"Address: {self.location}")


# Sample Input
order = DeliveryInfo("Suman", "Hyderabad")
order.display_delivery()
```
