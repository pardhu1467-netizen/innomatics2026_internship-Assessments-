```python

# Task 1: Find Employees with Highest Score


def find_top_employees(score_data):
    # Find highest score from dictionary values
    highest_score = max(score_data.values())
    
    # Store employees who have highest score
    top_employees = []
    
    for name, score in score_data.items():
        if score == highest_score:
            top_employees.append(name)
    
    return highest_score, top_employees


# Sample employee data
employee_scores = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Calling function
score, employees = find_top_employees(employee_scores)

print(f"Employees eligible for bonus: {', '.join(employees)} (Score: {score})")
```

```python

# Task 2: Count Frequency of Products

def count_products(product_list):
    # Clean product names (remove spaces and convert to lowercase)
    cleaned = [item.strip().lower() for item in product_list]
    
    frequency = {}
    
    # Count each product occurrence
    for product in cleaned:
        if product in frequency:
            frequency[product] += 1
        else:
            frequency[product] = 1
            
    return frequency


products = ["Phone", "Laptop", "Tablet", "  Phone", "Laptop", "Headset ", "Tablet"]

result = count_products(products)

print("Product Frequency:", result)
```

```python

# Task 3: Separate Even and Odd Readings
def classify_readings(values):
    even_readings = []
    odd_readings = []
    
    # Check each value
    for index in range(len(values)):
        if values[index] % 2 == 0:
            even_readings.append((index, values[index]))
        else:
            odd_readings.append((index, values[index]))
            
    return even_readings, odd_readings


sensor_values = [3, 4, 7, 8, 10, 12, 5]

even_data, odd_data = classify_readings(sensor_values)

print("Even Readings:", even_data)
```

```python

# Task 4: Calculate Email Domain Percentage

def calculate_domain_percentage(emails):
    gmail_count = 0
    yahoo_count = 0
    
    # Count domain occurrences
    for email in emails:
        if "@gmail.com" in email:
            gmail_count += 1
        elif "@yahoo.com" in email:
            yahoo_count += 1
    
    total = len(emails)
    
    gmail_percentage = (gmail_count / total) * 100
    yahoo_percentage = (yahoo_count / total) * 100
    
    return gmail_percentage, yahoo_percentage


email_list = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

gmail_per, yahoo_per = calculate_domain_percentage(email_list)

print("Gmail Usage: {:.2f}%".format(gmail_per))
print("Yahoo Usage: {:.2f}%".format(yahoo_per))
```

```python

# Task 5: Display Above Average Sales
def display_above_average(sales):
    # Calculate average
    average_sales = sum(sales) / len(sales)
    print("Average Sales:", average_sales)
    
    print("Days with Above Average Sales:")
    
    day = 1
    for amount in sales:
        if amount > average_sales:
            print(f"Day {day}: {amount}")
        day += 1


weekly_sales = [1200, 1500, 900, 2200, 1400, 3000]

display_above_average(weekly_sales)
```

```python

# Task 6: Find Duplicate User IDs

def find_duplicate_users(users):
    count_tracker = {}
    
    # Count occurrences of each user
    for user in users:
        count_tracker[user] = count_tracker.get(user, 0) + 1
    
    duplicates = {}
    
    # Extract users appearing more than once
    for user, count in count_tracker.items():
        if count > 1:
            duplicates[user] = count
            
    return duplicates


user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

duplicate_users = find_duplicate_users(user_ids)

print("Duplicate Users Found:")
for user, times in duplicate_users.items():
    print(f"{user} -> {times} times")
```
